## Distance matrix visualization

```python
# AWS
drive_path = '/data/dynamic-brain-workshop/brain_observatory_cache/'
import numpy as np
import pandas as pd
import os
import sys
import h5py

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sds
```

```python
from sklearn.manifold import Isomap, LocallyLinearEmbedding, SpectralEmbedding, TSNE, MDS
```

Grab data

```python
test_data = pd.read_hdf('Sample_array.h5')
```

```python
sds.heatmap(test_data)
```

Get projection to D2

```python
mds = MDS(n_components=2, dissimilarity="precomputed", random_state=6)
results = mds.fit(1-test_data.values)
```

```python
coords = results.embedding_

plt.subplots_adjust(bottom = 0.1)
plt.scatter(
    coords[:, 0], coords[:, 1], marker = 'o'
    )

for label, x, y in zip(test_data.columns, coords[:, 0], coords[:, 1]):
    plt.annotate(
        label,
        xy = (x, y), xytext = (-20, 20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
```
