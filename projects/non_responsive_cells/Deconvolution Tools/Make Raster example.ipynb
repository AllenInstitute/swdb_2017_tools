```python
import numpy as np
import pandas as pd
import os
import sys
import h5py

import matplotlib.pyplot as plt
%matplotlib inline

drive_path = '/Volumes/Brain2017/data/dynamic-brain-workshop/brain_observatory_cache/'

from allensdk.core.brain_observatory_cache import BrainObservatoryCache

manifest_file = os.path.join(drive_path,'brain_observatory_manifest.json')

boc = BrainObservatoryCache(manifest_file=manifest_file)

expt_container_id=511510736 #example exp container
expt_session_info = boc.get_ophys_experiments(experiment_container_ids=[expt_container_id])

ophys_experiment_id = 501474098 #example experiment
dataset = boc.get_ophys_experiment_data(ophys_experiment_id=ophys_experiment_id)

dff_traces = dataset.get_dff_traces()[1]
timestamps = dataset.get_fluorescence_timestamps()
```

```python
from OASIS.functions import deconvolve  #https://github.com/j-friedrich/OASIS
```

```python
timestamps = np.asarray(timestamps)
```

```python
spike_times = np.zeros(dff_traces.shape, dtype=object)
spike_times.shape

spike_times = []
isis = []
spikes = []
for i, dff_trace in enumerate(dff_traces):
    c, s, b, g, lam = deconvolve(np.double(dff_trace), penalty=1)
    s_sig = (s >= 3*np.std(s))
    spikes.append(s_sig*1)
    spike_times.append(timestamps[s_sig])
    isis.append(np.diff(spike_times[-1]))    
```

```python
plt.figure(num=None, figsize=(16, 6), dpi=80, facecolor='w', edgecolor='k')
for i, spike_time in enumerate(spike_times):
    plt.scatter(spike_time,np.ones(len(spike_time))+i, s=0.2, c='k', marker='o')
plt.show()
   
```

```python
import csv

with open('isis.csv', 'wb') as myfile:
    wr = csv.writer(myfile)
    [wr.writerow(line) for line in isis]
    
with open('spike_times.csv', 'wb') as myfile:
    wr = csv.writer(myfile)
    [wr.writerow(line) for line in spike_times]

with open('spikes.csv', 'wb') as myfile:
    wr = csv.writer(myfile)
    [wr.writerow(line) for line in spikes]
```
